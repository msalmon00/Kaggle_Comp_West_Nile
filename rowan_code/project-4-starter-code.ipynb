```python
weather = pd.read_csv("../assets/weather.csv")
weather.head()
```

```python
weather = weather[weather.Station == 1]
```

```python
weather.drop(["CodeSum","Depth","Station", "SnowFall", "Water1"], axis=1, inplace=True)
```

```python
def eda(dataframe): #Ritkas code chunk to check quality of data
    print "missing values \n", dataframe.isnull().sum() #shows total amount of null values for each column
    print "dataframe types \n", dataframe.dtypes
    print "dataframe shape \n", dataframe.shape     
    print "dataframe describe \n", dataframe.describe()
    print "dataframe length =", len(dataframe) #length of the dataframe
    print "duplicates", dataframe.duplicated().sum() # added this to duplicates in the data
    for item in dataframe:
        print item
        print dataframe[item].nunique()
        
eda(weather)
```

```python
weather[weather['Tavg'] == 'M']
weather.ix[weather.Tavg == "M", "Tavg"] = ((weather.Tmax + weather.Tmin)/2)
#Mike's piece of code for imputing missing values, this time replacing missing values in Tavg 
# with the average of TMax and TMin
```

```python
weather.ix[weather.WetBulb == "M", "WetBulb"] = ((weather.Tavg + weather.DewPoint)/2)
```

```python
weather.Date = pd.to_datetime(weather["Date"])
weather['Tavg']= weather['Tavg'].astype('float')
weather['Heat']= weather['Heat'].astype('float')
weather['Cool']= weather['Cool'].astype('float')
weather['Depart']= weather['Depart'].astype('float')
weather['WetBulb']= weather['WetBulb'].astype('float')
weather['SeaLevel']= weather['SeaLevel'].astype('float')
weather['StnPressure']= weather['StnPressure'].astype('float')
```

```python
weather.PrecipTotal = weather.PrecipTotal.str.replace(r'T', '0').astype(float)
```
